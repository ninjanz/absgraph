```python
import absgraph.utils.abstract_helper as abs_h
import absgraph.utils.converter as abs_c
import absgraph.utils.server_funcs as abs_s
import absgraph.utils.stats as abs_st

from collections import defaultdict

import matplotlib.pyplot as py
import numpy as np

import json
import pymongo
import time
```

```python
mongo = pymongo.MongoClient()
mng_cols = mongo['arguments'].collection_names()
```

```python
# first convert all graphs to abstract form

for col in mng_cols:
    # create a cursor that will loop thorugh all the docs in a collection
    cursor = mongo['arguments'][col].find()
    
    # now iterate through the cursor
    for doc in cursor:
        new_doc = {}
        new_doc['_id'] = doc['_id']
        print('--------------------------------------------------------\n' +
              doc['_id'] + 
              '\n--------------------------------------------------------')
        
        # list of nodes and string to input to server
        nodes_list = abs_h.map_nodes(doc)
        input_str = abs_h.rename_nodes(abs_h.get_web_input(doc, nodes_list), nodes_list)
        
        if input_str is not None:
            # convert to abstract using web service
            abs_form = abs_s.convert_abstract(input_str)
            
            if abs_form is not None:
                print(abs_form)
                # generate all the stats
                eval_form = abs_s.evaluate(input_str)
                
                abs_ = json.loads(abs_form)
                eval_ = json.loads(eval_form)
                instr_ = json.loads(input_str)
                
                new_doc.update(abs_st.stats_counter(abs_, eval_, instr_))
                
                new_doc['abs'] = abs_
                new_doc['eval'] = eval_
                new_doc['converted'] = abs_c.convert(abs_)
                
                print(new_doc)
                print('||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||\n')
                
                mongo['abstract']['tgf_stats'].replace_one({'_id': doc['_id']}, new_doc, upsert=True)
```

```python
# collecting all the information

all_stats = defaultdict(list)
cursor = mongo['abstract']['tgf_stats'].find()

print('-----------------------\n' +
      'total docs: ' + str(cursor.count()) + 
      '\n-----------------------')

keys = ['n_nodes', 'n_edges', 'n_cycles', 'n_rules', 'n_ext', 'density']

for doc in cursor:
    for k in doc.keys():
        if k in keys:
            all_stats[k].append(doc[k])
            
        elif k == 'avg_eval_time':
            all_stats['local_eval'].append(doc[k][0])
            all_stats['web_eval'].append(doc[k][1])
            
        elif k == 'n_contrary':
            all_stats['contradictory'].append(doc[k][0])
            all_stats['contrary'].append(doc[k][1])
```
